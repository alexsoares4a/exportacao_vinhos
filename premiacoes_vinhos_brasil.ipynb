```python
import pandas as pd

# Importa a função unidecode para remover acentos de strings
from unidecode import unidecode
```

```python
url = "dados/premiacoes_vinhos_brasil.xlsx"
url_paises_continentes = "dados/paises_continentes.csv"

dados_premiacoes = pd.read_excel(url)
dados_paises_continentes = pd.read_csv(url_paises_continentes, sep=";")

dados_premiacoes.head(5)
```

```python
dados_premiacoes.info()
```

```python
dados_premiacoes["Países Participantes"] = dados_premiacoes["Países Participantes"].fillna(0)
dados_premiacoes["Países Participantes"].astype(int)
```

```python
# Remover acentos
dados_premiacoes["País do Evento"] = dados_premiacoes["País do Evento"].apply(unidecode)

dados_premiacoes.head(15)
```

```python
dados_paises_continentes
```

```python
dados_premiacoes = pd.merge(dados_premiacoes, dados_paises_continentes, how='left', left_on='País do Evento', right_on='pais')
```

```python
dados_premiacoes.to_csv("dados_avaliacoes_dados_premiacoes.csv", sep=";", encoding="ISO-8859-1")
```

# **Total Prêmios por Ano**

```python
dados_avaliacoes_total_premios = dados_premiacoes.groupby(["Ano"])[["Total de Prêmios"]].sum("Total de Prêmios")
dados_avaliacoes_total_premios
```

## **Total Prêmios entre 2011 e 2023**

```python
dados_avaliacoes_total_premios.drop([2010, 2024], inplace=True)

dados_avaliacoes_total_premios
```

```python
dados_avaliacoes_total_premios.to_csv("dados_avaliacoes_total_premios.csv", sep=";", encoding="ISO-8859-1")
```

```python
dados_premiacoes[(dados_premiacoes["País do Evento"] != "Brasil")]
```

```python
dados_avaliacoes_sem_brasil = dados_premiacoes[(dados_premiacoes["País do Evento"] != "Brasil")]

dados_avaliacoes_sem_brasil_total_premios = dados_avaliacoes_sem_brasil.groupby("Ano")[["Total de Prêmios"]].sum("Total de Prêmios")

dados_avaliacoes_sem_brasil_total_premios.drop([2010, 2024], inplace=True)

dados_avaliacoes_sem_brasil_total_premios
```

```python
dados_avaliacoes_sem_brasil_total_premios.to_csv("dados_avaliacoes_sem_brasil_total_premios.csv", sep=";", encoding="ISO-8859-1")
```

# **Total Prêmios entre 2011 e 2023 por Países**

```python
dados_avaliacoes_por_pais = dados_avaliacoes_sem_brasil.groupby(["País do Evento", "continente"])[["Total de Prêmios"]].sum("Total de Prêmios").sort_values("Total de Prêmios", ascending=False)

dados_avaliacoes_por_pais.to_csv("dados_avaliacoes_por_pais.csv", sep=";", encoding="ISO-8859-1")
```

```python
dados_avaliacoes_por_continente = dados_avaliacoes_sem_brasil.groupby(["continente"])[["Total de Prêmios"]].sum("Total de Prêmios").sort_values("Total de Prêmios", ascending=False)

dados_avaliacoes_por_continente.to_csv("dados_avaliacoes_por_continente.csv", sep=";", encoding="ISO-8859-1")
```
